```python
'''#1.Given the dictionary'test_results',your task is to iterate over the
values of the dictionary and print all names of people who received less than
45 points.'''
test_results={'Alice':50,'Bob':45,'Charlie':30,'David':40,'Eve':50,'Fiona':45,'George':
20}
for i in test_results:
    if test_results[i]<45:
        print(i,test_results[i])

#OR

for x,y in test_results.items():
    if y<45:
        print(x)
        print(y)
```

```python
'''#2. Write a python program that takes a`frequencies`dictionary and a list of
vowels and prints the total number of consonants and the total number of vowels'''.
vowels=['a','e','i','u','o']
frequencies={'a':2,'s':1,'i':3,'u':3,'m':4,'l':2,
             'z':1,'n':3,'k':1,'e':4,'y':2,'d':2,'o':1}
v={}
c={}
for i in frequencies:
    if i in vowels:
        v[i]=frequencies[i]
    else:
        c[i]=frequencies[i]
print(c)
print(v)
a=sum(v.values())
print(a)
b=sum(c.values())
print(b)


#OR

vcount=0
ccount=0
for i in frequencies:
    if i in vowels:
        vcount+=frequencies[i]
    else:
        ccount+=frequencies[i]
print(vcount)
print(ccount)
```

```python
'''Q3.Write a python program that prints all restricted words from a given
word_list`one by one in separate lines.
A word is considered restricted if it has at least one character from the
`special_chars` list.'''
special_chars=['!','@','#','$','%','^','&','*']
word_list=['se#$et','hello_world','Ver*%&$lo',
           'di$#^$nt','c*%e','is','#%$#@!@#$%^$#']
for i in word_list:
    for j in special_chars:
        if j in i:
            print(i)
            break
```

```python
'''Q5.Paul and Tina each have a list of the irrespective friends:
paul_friends=["Grace","Bob","Cal","Dave"]
tina_friends=["Eric","Dave","Grace","Hugo"]
a.Combine both lists into a single list that contains all of their friends.
Donâ€™t include duplicate entries in the resulting list.
b.Now,use a for loop to get a list of their common friends.'''

paul_friends=["Grace","Bob","Cal","Dave"]
tina_friends=["Eric","Dave","Grace","Hugo"]
s1=set(paul_friends)
s2=set(tina_friends)
s3=s1.union(s2)
print(s3)
commonfriends=[]
for i in paul_friends:
    if i in tina_friends:
        commonfriends.append(i)
print(commonfriends)
```

```python
'''Q6.You work at a sports club.The following sets contain the names of players
registered to play different sports:
golf_players={"Evan","Tom","Tim","Alex"}
volleyball_players={"Jack","Hugh","Alex","Martyn"}
cricket_players={"Evan","Tim","Jessica","Martyn","Michael"}
How can you obtain a set that includes the players that are only registered
to play cricket (i.e.,not registered for golf or volleyball)?'''

golf_players={"Evan","Tom","Tim","Alex"}
volleyball_players={"Jack","Hugh","Alex","Martyn"}
cricket_players={"Evan","Tim","Jessica","Martyn","Michael"}
onlycricketplayers=set()
for i in cricket_players:
    if i not in golf_players and i not in volleyball_players:
        onlycricketplayers.add(i)
print(onlycricketplayers)
```

```python
'''Q7.Given the list of words below...
#zodiac_signs=['taurus','gemini','cancer','leo','virgo','libra',
#'scorpio','sagittarius','capricorn','aquarius','pisces']
#Find the longest and the shortest word in the list.'''

zodiac_signs=['taurus','gemini','cancer','leo','virgo','libra',
'scorpio','sagittarius','capricorn','aquarius','pisces']
shortest_word=zodiac_signs[0]
longest_word=zodiac_signs[0]
for i in zodiac_signs:
    if len(i)>len(longest_word):
        longest_word=i
    elif len(i)<len(shortest_word):
        shortest_word=i
print(shortest_word)
print(longest_word)
```
